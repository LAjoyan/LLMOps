### Clean email json file
Format that MLFlow requires:

- input -> inputs
- {"expected_response": '{'sender_name': 'Erik Lindqvist', 'sender_emai...'}'

```python
df = df.rename({'input': 'inputs'}, axis=1)

df
```

```python
df['expectations'].apply(lambda row: {'expected_response': row})
```

```python
import json
df['expectations'] = df['expectations'].apply(lambda row: {'excepted_response':json.dumps(row)})
df
```

```python
df.to_json('emails_cleaned.json', index=False, orient='records')  # Convert DataFrame to JSON file (list of records, no index)
```
