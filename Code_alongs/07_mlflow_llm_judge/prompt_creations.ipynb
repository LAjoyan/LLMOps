## Prompt creation and versioning

```python
from mlflow.genai import register_prompt

email_summary_description =register_prompt(
    name='summary-description',
    template='summarize in 3-4 sentences',
    commit_message='Initail summary description prompt',
    tags={'author': 'Lilit',
          'agent': 'email_extractor',
          'prompt_type': 'description',}
)

email_summary_description
```

```python
email_summary_description =register_prompt(
    name='summary-description',
    template='summarize in {{num_sentences}} sentences, use same langauge as original text',
    commit_message='Variable sentences and language ',
    tags={'author': 'Lilit',
          'agent': 'email_extractor',
          'prompt_type': 'description',}
)

email_summary_description
```

## Load a prompt

```python
from mlflow.genai import load_prompt

email_summary_description = load_prompt("summary-description", version=2)
email_summary_description
```

```python
email_summary_description.format(num_sentences=4)
```
